# <span style='font-family:"Times New Roman"'>Generating and Scaling Gaussian Noise With 0 mean and standard deviation of 64:

```python
mu, sigma = 0, 64 # mean and standard deviation
noise = np.random.normal(mu, sigma, (1548,2238))
noise = noise - np.amin(noise)
noise = (noise/(np.amax(noise))) * 255
```

```python
smallestn = np.amin(noise)
biggestn = np.amax(noise)
```

```python
print("Minimum value is:")
print(smallestn)
print("Maximum value is:")
print(biggestn)
```

# <span style='font-family:"Times New Roman"'> Adding Gaussian Noise to Image and Scaling it:

```python
Noisyimage = FI + noise
Noisyimage = Noisyimage - np.amin(Noisyimage)
Noisyimage = (Noisyimage/(np.amax(Noisyimage))) * 255
```

# <span style='font-family:"Times New Roman"'>Averaging process for 1000 images:

<span style='font-family:"Times New Roman"'>At each loop, new random Gaussian noise would be generated and would be added to image. Then resulting image would be scaled. 

```python
FIN = []
for i in range(0, 1000):
    noise = np.random.normal(mu, sigma, (1548,2238))
    noise = noise - np.amin(noise)
    noise = (noise/(np.amax(noise))) * 255
    W = FI + noise
    W = W - np.amin(W)
    W = (W/(np.amax(W))) * 255
    FIN.append(W) 
```

```python
Noiselessimage1000 = sum(FIN)/len(FIN)
Noiselessimage1000 = Noiselessimage1000 - np.amin(Noiselessimage1000)
Noiselessimage1000 = (Noiselessimage1000/(np.amax(Noiselessimage1000))) * 255
```

# <span style='font-family:"Times New Roman"'> Displaying the resulting images:

```python
fig = plt.figure(figsize=(18, 16))
plt.subplot(3, 3, 1)
plt.imshow(Noisyimage, cmap='gray')
plt.title('Image with random noise')
plt.axis('off')
plt.subplot(3, 3, 2)
plt.imshow(Noiselessimage10, cmap='gray')
plt.title('Number of Averaged Image = 10')
plt.axis('off')
plt.subplot(3, 3, 3)
plt.imshow(Noiselessimage50, cmap='gray')
plt.title('Number of Averaged Image = 50')
plt.axis('off')
plt.subplot(3, 3, 4)
plt.imshow(Noiselessimage100, cmap='gray')
plt.title('Number of Averaged Image = 100')
plt.axis('off')
plt.subplot(3, 3, 5)
plt.imshow(Noiselessimage500, cmap='gray')
plt.title('Number of Averaged Image = 500')
plt.axis('off')
plt.subplot(3, 3, 6)
plt.imshow(Noiselessimage1000, cmap='gray')
plt.title('Number of Averaged Image = 1000')
plt.axis('off')
fig.suptitle('Effect of Averaging for Noise Cancellation ', fontsize = 20)
plt.show()
```

<span style='font-family:"Times New Roman"'>The essential part of averaging method that cancels the noise is the zero mean of Gaussian noise. Since the mean is zero, when added to noiseless image, it is not going to change the mean of resulting image. Now, by repeating the procedure, we are going to have different pictures with different noise added to them with same standard deviation and same mean. So, as we repeat this addition with large numbers and averaging it, we are going to get closer to real values thanks to zero mean of gaussian noise.
Following simple example will make the description easier. 
Imagine we have x = 1. Now, we are adding and subtracting small numbers randomly at each step and storing it. For instance, we are adding 0.01 at first step and we have 1.01. then we subtract 0.2 from one and we have 0.8. The mean of 0.8 and 1.01 is 0.808. as the repetition of this process increase and numbers fluctuate more around one, the mean would tend to get closer to 1.   
<span style='font-family:"Times New Roman"'>This is how averaging works for noise reduction.    

# <span style='font-family:"Times New Roman"'> Calculating Mean Square Error:

```python
msenoisy = (np.square(FI-Noisyimage)).mean(axis=None)
mse10 = (np.square(FI-Noiselessimage10)).mean(axis=None)
mse50 = (np.square(FI-Noiselessimage50)).mean(axis=None)
mse100 = (np.square(FI-Noiselessimage100)).mean(axis=None)
mse500 = (np.square(FI-Noiselessimage500)).mean(axis=None)
mse1000 = (np.square(FI-Noiselessimage1000)).mean(axis=None)
```

```python
x = [0,10,50,100,500,1000]
y = [msenoisy,mse10,mse50,mse100,mse500,mse1000]
```

# <span style='font-family:"Times New Roman"'>Displaying MSE values with respect to number of averaging: 

```python
fig = plt.figure(figsize=(14, 10))
ax = plt.axes()
ax.plot(x, y);
plt.title('Effect of Image Averaging on MSE of Noisy Image Compared to Noisless Image')
plt.xlabel('Number of Image Averaging',fontsize=18)
plt.ylabel('MSE',fontsize=18)
plt.show()
```

<span style='font-family:"Times New Roman"'>The plot asserts that by increasing the number of averaging, the mean square error is decreasing. Meaning that increasing the averaging time would result in better noise canceling.
