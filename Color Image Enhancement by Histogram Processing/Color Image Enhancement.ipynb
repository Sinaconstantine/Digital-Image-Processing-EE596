### <span style='font-family:"Times New Roman"'>This problem will study the effects of histogram-based processing on color images by looking at the results when histogram equalization is applied in the RGB versus the HSI color space.

# <span style='font-family:"Times New Roman"'>Loading Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import exposure
from skimage import color
```

# <span style='font-family:"Times New Roman"'>Reading and Displaying Image

```python
I_rgb = ((mpimg.imread('dark_stream.tif')).astype(np.uint8))
img=plt.imshow(I_rgb)
plt.axis('off')
plt.title('Dark Stream Image', fontsize = 16)
plt.show()
```

```python
R = I_rgb[:,:,0]
G = I_rgb[:,:,1]
B = I_rgb[:,:,2]
R_histeq = exposure.equalize_hist(R)
G_histeq = exposure.equalize_hist(G)
B_histeq = exposure.equalize_hist(B)
I_rgb_histeq = np.zeros((697,697,3))
I_rgb_histeq[:,:,0] = R_histeq
I_rgb_histeq[:,:,1] = G_histeq
I_rgb_histeq[:,:,2] = B_histeq
```

```python
img=plt.imshow(I_rgb_histeq)
plt.axis('off')
plt.title('Histogram Equalized Dark Stream Image', fontsize = 16)
plt.show()
```

```python
I_hsv=color.rgb2hsv(I_rgb)
```

```python
H = I_hsv[:,:,0]
S = I_hsv[:,:,1]
V = I_hsv[:,:,2]
H_histeq = exposure.equalize_hist(H)
S_histeq = exposure.equalize_hist(S)
V_histeq = exposure.equalize_hist(V)
I_hsv_histeq = np.zeros((697,697,3))
I_hsv_histeq[:,:,0] = H
I_hsv_histeq[:,:,1] = S
I_hsv_histeq[:,:,2] = V_histeq
```

```python
img=plt.imshow(I_hsv_histeq)
plt.axis('off')
plt.title('I_hsv_histeq Image', fontsize = 16)
plt.show()
```

<span style='font-family:"Times New Roman"'>After equalizing value in HSV system, for accurate representation, we should transfer each channel back to RGB and display them. Actually, the HSV or HSI domain are better domain for color histogram equalization by keeping hue unchanged and spreading intensity values. 

```python
I_hsv_histeq_rgb=color.hsv2rgb(I_hsv_histeq)
```

```python
img=plt.imshow(I_hsv_histeq_rgb)
plt.axis('off')
plt.title('Resulting Image from Converting Equalized HSV Image to RGB', fontsize = 16)
plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
plt.imshow(I_rgb_histeq)
plt.title('I_rgb_histeq', fontsize = 16)
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(I_hsv_histeq)
plt.title('I_hsv_histeq', fontsize = 16)
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(I_hsv_histeq_rgb)
plt.title('I_hsv_histeq_rgb', fontsize = 16)
plt.axis('off')
fig.suptitle('Effect of Equalizing Different Channels of Dark Stream Image', fontsize = 20)
plt.show()
```

<span style='font-family:"Times New Roman"'> As it is mentioned in the book, it is unwise to equalize each channel of the image and display them. As a result, in this problem, although RGB equalized image may seem fine, the colors are not vibrant compared to case when we have equalized the V channel in HSV system and convert it back to RGB. The resulted equalized image in HSV system is much better with keeping the details and having good color representation. 
HSV or HSI are better domains to equalize the image (just V or I channels). 
The I_hsv_histeq is a wrong displaying method and it is not comparable with other two images.
