### <span style='font-family:"Times New Roman"'>In this program, grayscale imageâ€™s intensity level would be manipulated. 

# <span style='font-family:"Times New Roman"'>Loading Required Modules:

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

#<span style='font-family:"Times New Roman"'> Reading The Image:

```python
im = np.array(Image.open('drip-bottle.tif'))
```

#<span style='font-family:"Times New Roman"'> Let's have a look at data properties: 

```python
print("Data type is:")
print(im.dtype)
print("Data dimension is:")
print(im.ndim)
print("Number of image's rows and columns:")
print(im.shape)
```

# <span style='font-family:"Times New Roman"'>Changing intensity Levels:

```python
im_2 = im // 128 * 128
im_4 = im // 64 * 64
im_8 = im // 32 * 32
im_16 = im // 16 * 16
im_32 = im // 8 * 8
im_64 = im // 4 * 4
im_128 = im // 2 * 2
```

#<span style='font-family:"Times New Roman"'> Plotting resulted images:

```python
fig = plt.figure(figsize=(18, 16))
plt.subplot(4, 4, 1)
plt.imshow(im, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 256')
plt.axis('off')
plt.subplot(4, 4, 2)
plt.imshow(im_128, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 128')
plt.axis('off')
plt.subplot(4, 4, 3)
plt.imshow(im_64, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 64')
plt.axis('off')
plt.subplot(4, 4, 4)
plt.imshow(im_32, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 32')
plt.axis('off')
plt.subplot(4, 4, 5)
plt.imshow(im_16, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 16')
plt.axis('off')
plt.subplot(4, 4, 6)
plt.imshow(im_8, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 8')
plt.axis('off')
plt.subplot(4, 4, 7)
plt.imshow(im_4, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 4')
plt.axis('off')
plt.subplot(4, 4, 8)
plt.imshow(im_2, cmap='gray', vmin=0, vmax=255)
plt.title('Intensity Level : 2')
plt.axis('off')
fig.suptitle('Drip bottle image with different intensity levels', fontsize = 20)
plt.show()
```
