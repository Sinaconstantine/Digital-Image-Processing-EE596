```python
I_scaled=(I/2).astype(np.uint8)
```

```python
img=plt.imshow(I_scaled) 
plt.axis('off')
```

```python
img=plt.imshow(I_scaled)
plt.axis('auto')
```

<span style='font-family:"Times New Roman"'>In contrast to displaying gray scale image, the two commands here are showing similar result. The reason lies upon the type of data. When working with gray scale image, python assign color to image. But in color image, the color is already there and python cannot manipulate the colors.

<span style='font-family:"Times New Roman"'>I_scaled and I has the same data types, same size, and same dimension. However, by dividing I by 2, we are changing the intensity values in the image, so they become smaller. As we expect, the resulting image should be darker compared to real image since it has smaller intensity values.  

```python
I_float = I.astype(float)
```

```python
img=plt.imshow(I_float)
plt.axis('off')
```

```python
img=plt.imshow(I_float)
plt.axis('auto')
```

<span style='font-family:"Times New Roman"'>When casting the image in float, the result look like negative results. The reason behind this fact is simple. In requested code, for displaying the image properly as float, we need to normalize data range between 0 and 1. 
The resulting image with both type of representation is similar. 

```python
img=plt.imshow(I_float/255)
plt.axis('auto')
```

```python
R = I[:,:,0]
G = I[:,:,1]
B = I[:,:,2]
```

```python
fig = plt.subplots(figsize=(18, 10))
plt.subplot(2, 3, 1)
plt.imshow(R, cmap=plt.get_cmap('gray'))
plt.title('Red channel', fontsize = 20)
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(G,cmap=plt.get_cmap('gray'))
plt.title('Green channel', fontsize = 20)
plt.axis('off')
plt.subplot(2, 3, 3)
plt.imshow(B,cmap=plt.get_cmap('gray'))
plt.title('Blue channel', fontsize = 20)
plt.axis('off')
plt.suptitle('Representation of image with one color channel', fontsize = 20)
plt.show()
```
