```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from skimage import filters
import numpy as np
```

```python
I = mpimg.imread('text-spotshade.tif')
```

```python
Length = 20
constant = 0.5
I_thresh = np.zeros(I.shape)
m = np.ones(I.shape)
for r in range (I.shape[0]):
    for c in range(1,I.shape[1]):
        m[:,c-1] = I[:,c-1]
        m[r,c] = m[r,c-1] + 1/Length * (I[r,c]-I[r,c-Length])
T = constant * m
I_thresh[I>T] = 1
```

```python
T_otsu=filters.threshold_otsu(I)
result = I > T_otsu
```

```python
fig = plt.figure(figsize=(16, 10))
plt.subplot(1, 3, 1)
img=plt.imshow(I,cmap = 'gray')
plt.title('Original Image', fontsize = 16)
plt.axis('off')
plt.subplot(1, 3, 2)
img=plt.imshow(result,cmap='gray')
plt.axis('off')
plt.title('Otsu Thresholding Image',fontsize=16)
plt.subplot(1, 3, 3)
plt.imshow(I_thresh,cmap = 'gray')
plt.title('Variable Thresholding Based on Moving Average', fontsize = 16)
plt.axis('off')
plt.show()
```
