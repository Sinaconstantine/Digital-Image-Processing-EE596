### <span style='font-family:"Times New Roman"'>This problem will study segmentation of color images based on distance in a color space

# <span style='font-family:"Times New Roman"'>Loading Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# <span style='font-family:"Times New Roman"'>Reading and Displaying Image

```python
fig = plt.figure(figsize=(8, 8))
I_rgb = ((mpimg.imread('strawberries-RGB.tif')).astype(np.uint8))
img=plt.imshow(I_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
I_cube = I_rgb/255
W = 0.2549
a1, a2, a3 = 0.6863, 0.1608, 0.1922
```

# <span style='font-family:"Times New Roman"'> Distance-Based Color Slicing Equation 7-43 Gonzalez and Wood Book

```python
I_cube[(np.absolute(I_cube[:,:,0]- a1) > W/2)] = 0.5
I_cube[(np.absolute(I_cube[:,:,1]- a2) > W/2)] = 0.5
I_cube[(np.absolute(I_cube[:,:,2]- a3) > W/2)] = 0.5
```

```python
I_sph = I_rgb/255
R = 0.1765
```

# <span style='font-family:"Times New Roman"'> Distance-Based Color Slicing Equation 7-44 Gonzalez and Wood Book

```python
I_sph[((I_sph[:,:,0] - a1)**2 + (I_sph[:,:,1] - a2)\
       **2 + (I_sph[:,:,2] - a3)**2) > R ** 2] = 0.5
```

```python
fig = plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.imshow(I_cube)
plt.title('Cubic Transformation', fontsize = 16)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(I_sph)
plt.title('Spherical Transformation', fontsize = 16)
plt.axis('off')
fig.suptitle('Color-Slicing Transformation for Red Color', fontsize = 20)
plt.show()
```
