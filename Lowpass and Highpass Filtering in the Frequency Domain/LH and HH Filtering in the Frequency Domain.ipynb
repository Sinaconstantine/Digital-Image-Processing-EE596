# <span style='font-family:"Times New Roman"'>Lowpass and Highpass Filtering in the Frequency Domain

# <span style='font-family:"Times New Roman"'>Loading Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math
```

```python
fig = plt.figure(figsize=(18, 6))
I = mpimg.imread('characterTestPattern688.tif')
imgplot = plt.imshow(I, cmap = 'gray')
plt.title('Original Image', fontsize = 20)
plt.axis('off')
plt.show()
```

```python
n = int(I.shape[0]/2)                #Pad Size
I_padded = np.pad(I, ((n,n),(n,n)), 'constant', constant_values = 0)
```

```python
fig = plt.figure(figsize=(18, 6))
imgplot = plt.imshow(I_padded, cmap = 'gray')
plt.title('Zero Padded Image', fontsize = 20)
plt.axis('off')
plt.show()
```

```python
NCFI = np.fft.fft2(I_padded)
CFI = np.fft.fftshift(NCFI)   # Centered Frequency
```

# <span style='font-family:"Times New Roman"'>Filter Design Function

```python
def gauss_lowpass(M,N,cm,cn,sigma):
    for i in range(M):
        for j in range(N):
            Dist = math.sqrt(((i-cm)**2)+((j-cn)**2))
            TF = np.exp(-(Dist**2)/(2*sigma**2))
            H[i,j] = TF
    return H
```

```python
M = int(I_padded.shape[0])
N = int(I_padded.shape[1])
cm = int(I_padded.shape[0]/2)
cn = int(I_padded.shape[1]/2) 
sigma = 10
H = np.zeros((1376,1376))
H10 = gauss_lowpass(M,N,cm,cn,sigma)
TI = CFI * H10
UCTI = np.fft.fftshift(TI)
FI = np.real(np.fft.ifft2(UCTI))
FI = np.delete(FI, range(n), axis=0)
FI = np.delete(FI, range(int(I.shape[0]), int(I.shape[0]) + n), axis=0)
FI = np.delete(FI, range(n), axis=1)
FI10 = np.delete(FI, range(int(I.shape[1]), int(I.shape[1]) + n), axis=1)
```
