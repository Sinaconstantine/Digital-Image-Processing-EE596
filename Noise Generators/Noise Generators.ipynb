### <span style='font-family:"Times New Roman"'>In this code, Gaussian, uniform, and salt and pepper noise will be generated and their effect on image histogram would be considered.  

#  <span style='font-family:"Times New Roman"'>Importing Required Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

#  <span style='font-family:"Times New Roman"'>Defining Functions 

```python
def add_gauss_noise(I,mu,sigma):
    row,col= I.shape
    gauss = np.random.normal(mu,sigma,(row,col))
    eta = gauss.reshape(row,col)
    g = I + eta
    return g, eta
```

```python
def add_uniform_noise(I,a,b):
    row,col= I.shape
    uniform = np.random.uniform(a,b,(row,col))
    eta = uniform.reshape(row,col)
    g = I + eta
    return g, eta
```

```python
def add_sp_noise(I,ps,pp):
    g = I.flatten()
    eta = np.zeros(I.size)
    indicesps = np.random.choice(np.arange(g.size), replace=False, size=int(g.size * ps))
    g[indicesps] = 0
    eta[indicesps] = 0
    indicespp = np.random.choice(np.arange(g.size), replace=False, size=int(g.size * pp))
    g[indicespp] = 255
    eta[indicesps] = 255
    return g, eta
```

```python
I = mpimg.imread('test-pattern.tif')
gg, eta = add_gauss_noise(I,0,20)
gg = gg - np.amin(gg)
gg = (gg/(np.amax(gg))) * 255
```

```python
fig = plt.subplots(figsize=(6, 6))
plt.imshow(I, cmap = 'gray')
plt.title('Original Image', fontsize = 20)
plt.axis('off')
fig = plt.subplots(figsize=(8, 6), facecolor = 'black')
plt.hist(I.flatten(),bins=255,range=(0,255),color = "white")
plt.title('Histogram of Original Image', fontsize = 20,color = "white")
plt.axis('off')
plt.show()
```

```python
fig = plt.subplots(figsize=(6, 6))
plt.imshow(gg, cmap = 'gray')
plt.title('Added Gaussian Noise', fontsize = 20)
plt.axis('off')
fig = plt.subplots(figsize=(8, 6), facecolor = 'black')
plt.hist(gg.flatten(),bins=255,range=(0,255),color = "white")
plt.title('Histogram of Image with Added Gaussian Noise', fontsize = 20,color = "white")
plt.axis('off')
plt.show()
```

```python
gu , eta = add_uniform_noise(I,0,75)
gu = gu - np.amin(gu)
gu = (gu/(np.amax(gu))) * 255
```

```python
fig = plt.subplots(figsize=(6, 6))
plt.imshow(gu, cmap = 'gray')
plt.title('Added Uniform Noise', fontsize = 20)
plt.axis('off')
fig = plt.subplots(figsize=(8, 6), facecolor = 'black')
plt.hist(gu.flatten(),bins=255,range=(0,255),color = "white")
plt.title('Histogram of Image with Added Uniform Noise', fontsize = 20,color = "white")
plt.axis('off')
plt.show()
```

```python
eta = []
gsp , eta= add_sp_noise(I,0.1,0.1)
gsp = np.reshape(gsp, I.shape)
```

```python
fig = plt.subplots(figsize=(6, 6))
plt.imshow(gsp, cmap = 'gray')
plt.title('Added Salt and Pepper Noise', fontsize = 20)
plt.axis('off')
fig = plt.subplots(figsize=(8, 6), facecolor = 'black')
plt.hist(gsp.flatten(),bins=255,range=(0,255),color = "white")
plt.title('Histogram of Image with Added Salt and Pepper Noise', fontsize = 20,color = "white")
plt.axis('off')
plt.show()
```

```python
fig = plt.subplots(figsize=(18, 10))
plt.subplot(2, 4, 1)
plt.imshow(I, cmap = 'gray')
plt.title('Original Image', fontsize = 20)
plt.axis('off')
plt.subplot(2, 4, 2)
plt.imshow(gg, cmap = 'gray')
plt.title('Added Gaussian Noise', fontsize = 20)
plt.axis('off')
plt.subplot(2, 4, 3)
plt.imshow(gu, cmap = 'gray')
plt.title('Added Uniform Noise', fontsize = 20)
plt.axis('off')
plt.subplot(2, 4, 4)
plt.imshow(gsp, cmap = 'gray')
plt.title('Added Salt and Pepper Noise', fontsize = 20)
plt.axis('off')
plt.suptitle('Effect of different noises on image and its histogram', fontsize = 20)
plt.show()
fig = plt.subplots(figsize=(18, 10), facecolor = 'black')
plt.subplot(2, 4, 1)
plt.hist(I.flatten(),bins=255,range=(0,255),color = "white")
plt.axis('off')
plt.subplot(2, 4, 2)
plt.hist(gg.flatten(),bins=255,range=(0,255),color = "white")
plt.axis('off')
plt.subplot(2, 4, 3)
plt.hist(gu.flatten(),bins=255,range=(0,255),color = "white")
plt.axis('off')
plt.subplot(2, 4, 4)
plt.hist(gsp.flatten(),bins=255,range=(0,255),color = "white")
plt.axis('off')
plt.show()
```
