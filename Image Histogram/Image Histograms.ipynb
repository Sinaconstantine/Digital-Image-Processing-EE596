# <span style='font-family:"Times New Roman"'>Loading required modules: 

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

# <span style='font-family:"Times New Roman"'>Reading image:

```python
I = mpimg.imread('pollen_dark.tif')
```

# <span style='font-family:"Times New Roman"'>Displaying image:

```python
img=plt.imshow(I,cmap='gray',vmin=0, vmax=255)
plt.axis('off')
plt.colorbar()
plt.title('pollen dark')
plt.show()
```

# <span style='font-family:"Times New Roman"'>Displaying image histogram: 

```python
fig = plt.figure(figsize=(10, 8))
plt.hist(I.flatten(),bins=255,range=(0,255))
plt.title('Histogram of loaded dark image', fontsize=20)
plt.xlabel('Intensity Values', fontsize=16)
plt.ylabel('Number of pixels', fontsize=16)
plt.show()
```

```python
h = np.histogram(I.flatten(),bins=256,range=(0,255))
```

 <span style='font-family:"Times New Roman"'>When using the command in above cell, two vectors will be returned. The first one is height of the values (numbers of specific values) and the other vector is the bins. 

```python
X = np.delete(h[1], 256)
height = h[0]
fig = plt.figure(figsize=(10, 8))
plt.bar(X,height)
plt.title('Histogram of loaded dark image using bar command', fontsize=20)
plt.xlabel('Intensity Values', fontsize=16)
plt.ylabel('Number of pixels', fontsize=16)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
plt.hist(I_float)
plt.title('Histogram of loaded dark image using plt.hist(I_float)', fontsize=20)
plt.xlabel('Intensity Values', fontsize=16)
plt.ylabel('Number of pixels', fontsize=16)
plt.show()
```

<span style='font-family:"Times New Roman"'>The histogram looks different from part (a). When casting image as a float, the intensity values remain the same, but float variable are more precise comparing to unit8. As a result, histogram is divided into precise intensity values and consequently, each individual bin has small number of pixels. The bins are in close adjacency of each other. Based on application and computational power, it can be a useful command. 

```python
fig = plt.figure(figsize=(10, 8))
plt.hist(I_float.flatten())
plt.title('Histogram of loaded dark image without defining bin number', fontsize=20)
plt.xlabel('Intensity Values', fontsize=16)
plt.ylabel('Number of pixels', fontsize=16)
plt.show()
```

<span style='font-family:"Times New Roman"'>The histogram looks different. Because when we do not specify number of bins, python considers random number of bins for histogram. That is why we have different histogram presentation. Setting the number of bins equal to 256 would result in same histogram.

```python
fig = plt.figure(figsize=(10, 8))
plt.hist(I_float.flatten(),bins=256,range=(0,255))
plt.title('Histogram of loaded dark image with defining bin number as 256' ,fontsize=20)
plt.xlabel('Intensity Values', fontsize=16)
plt.ylabel('Number of pixels', fontsize=16)
plt.show()
```

```python
I1 = mpimg.imread('pollen_dark.tif')
I4 = mpimg.imread('pollen_highcontrast.tif')
I2 = mpimg.imread('pollen_light.tif')
I3 = mpimg.imread('pollen_lowcontrast.tif')
```

```python
fig = plt.figure(figsize=(18, 16))
plt.subplot(4, 4, 1)
plt.imshow(I1,cmap='gray',vmin=0, vmax=255)
plt.title('Dark Image')
plt.axis('off')
plt.subplot(4, 4, 2)
plt.imshow(I2,cmap='gray',vmin=0, vmax=255)
plt.title('Light Image')
plt.axis('off')
plt.subplot(4, 4, 3)
plt.imshow(I3,cmap='gray',vmin=0, vmax=255)
plt.title('Low Contrast Iamge')
plt.axis('off')
plt.subplot(4, 4, 4)
plt.imshow(I4,cmap='gray',vmin=0, vmax=255)
plt.title('High Contrast Image')
plt.axis('off')
plt.subplot(4, 4, 5)
plt.hist(I1.flatten(),bins=256,range=(0,255))
plt.title('Histogram of Dark Image')
plt.subplot(4, 4, 6)
plt.hist(I2.flatten(),bins=256,range=(0,255))
plt.title('Histogram of Light Image')
plt.subplot(4, 4, 7)
plt.hist(I3.flatten(),bins=256,range=(0,255))
plt.title('Histogram of Low Contrast Image')
plt.subplot(4, 4, 8)
plt.hist(I4.flatten(),bins=256,range=(0,255))
plt.title('Histogram of High Contrast Image')
fig.suptitle('Four Image Types and Their Corresponding Histograms', fontsize = 20)
plt.show()
```
