# <span style='background:black'><span style='color:Cyan'> <span style='font-family:"Times New Roman"'> Objective: 
### <span style='font-family:"Times New Roman"'>In this program, using elementwise multiplication a mask for dental X-Ray image is designed. 

# <span style='font-family:"Times New Roman"'>Importing Required Modules 

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

# <span style='font-family:"Times New Roman"'>Reading the Image:

```python
I = mpimg.imread('dentalXray.tif')
```

```python
FI = I.astype(float)
```

# <span style='font-family:"Times New Roman"'>Creating a mask: 

```python
mask = np.zeros((674,882))
```

```python
mask[200:300, 300:550] = [255]
mask[400:520, 210:480] = [255]
```

# <span style='font-family:"Times New Roman"'>Element Wise multiplication: 

```python
MI = FI*mask
```

# <span style='font-family:"Times New Roman"'>Result: 

```python
fig = plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(FI, cmap='gray')
plt.title('Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title('Mask')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(MI, cmap='gray')
plt.title('Masked Image')
plt.axis('off')
fig.suptitle('Elementwise Matrix Multiplication Result on Images', fontsize = 20)
plt.show()
```

# <span style='font-family:"Times New Roman"'>Matrix multiplication:

```python
np.matmul(FI,mask)
```

<span style='font-family:"Times New Roman"'>The error from matrix multiplication is a dimensionality error. Since the image has 674x882 pixels, and the mask has 674x882 pixels, the matrix multiplication would not be possible since 882 is not equal to 674. For matrix multiplication, number of the columns for the first image should be equal to number of tows in second image.  

# <span style='font-family:"Times New Roman"'>Squaring the image and the mask: 

```python
SI = FI[100:600, 100:600]
SMI = mask[100:600, 100:600]
```

```python
Result = np.matmul(SI,SMI)
Result = Result - np.amin(Result)
Result = (Result/(np.amax(Result))) * 255
```

# <span style='font-family:"Times New Roman"'>Results: 

```python
fig = plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(SI, cmap='gray')
plt.title('Squared Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(SMI, cmap='gray')
plt.title('Squared Mask')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(Result, cmap='gray')
plt.title('Result of Matrix Multiplication')
plt.axis('off')
fig.suptitle('Matrix Multiplication Result on Images', fontsize = 20)
plt.show()
```

<span style='font-family:"Times New Roman"'>The drastic change in result of matrix multiplication comparing to element wise multiplication comes from the nature of matrix multiplication. The resulting value for each pixel, is sum of corresponding row to column elementwise multiplication. As a result, the resulting pixel is meaning less from image processing point of view.   
For displaying the image, scaling is essential since it would result in values larger than 255 and it should be scaled in order to be presentable. Also, as it is discussed before, squaring the images is important in order to perform matrix multiplication. 
In part 2, scaling was not required since the mask has intensity values of 0 or 1. Consequently, by performing element wise multiplication of image with mask, the range of resulting image would remain between 0 to 255. As a result, scaling was not required.
