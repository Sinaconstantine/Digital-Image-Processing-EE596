### <span style='font-family:"Times New Roman"'>Notch filter design for removing periodic noise  from saturn image. 

# <span style='font-family:"Times New Roman"'> Loading Required Package

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
fig = plt.figure(figsize=(18, 6))
I = mpimg.imread('suaturn-rings-sinusoidal-interf.tif')
imgplot = plt.imshow(I, cmap = 'gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
Magorg = np.fft.fft2(I)
magnitude= np.log(np.abs(np.fft.fftshift(Magorg)))
```

```python
fig = plt.figure(figsize=(10, 6))
img=plt.imshow(magnitude,cmap = 'gray')
plt.title('Magnitude Response of Original Image', fontsize = 16)
plt.axis('off')
plt.show()
```

```python
H = np.ones(I.shape)
H[0:674 ,332:342] = 0
H[320:354,:] = 1
```

```python
fig = plt.figure(figsize=(10, 6))
img=plt.imshow(H,cmap = 'gray')
plt.title('Notch Reject Filter', fontsize = 16)
plt.axis('off')
plt.show()
```

```python
NCFI = np.fft.fft2(I)
CFI = np.fft.fftshift(NCFI)   # Centered Frequency
TI = CFI * H
UCTI = np.fft.fftshift(TI)
FI = np.real(np.fft.ifft2(UCTI))
```

```python
fig = plt.figure(figsize=(10, 6))
img=plt.imshow(np.real(FI),cmap = 'gray')
plt.title('Filtered Image with Notch Reject Filter', fontsize = 16)
plt.axis('off')
plt.show()
```

```python
H1 = np.zeros(I.shape)
H1[0:674 ,332:342] = 1
H1[320:354,:] = 0
```

```python
fig = plt.figure(figsize=(10, 6))
img=plt.imshow(H1,cmap = 'gray')
plt.title('Notch Pass Filter', fontsize = 16)
plt.axis('off')
plt.show()
```

```python
NCFI = np.fft.fft2(I)
CFI = np.fft.fftshift(NCFI)   # Centered Frequency
TI = CFI * H1
UCTI = np.fft.fftshift(TI)
FI1 = np.real(np.fft.ifft2(UCTI))
```

```python
fig = plt.figure(figsize=(10, 6))
img=plt.imshow(np.real(FI1),cmap = 'gray')
plt.title('Filtered Image with Notch Pass Filter', fontsize = 16)
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(18, 14))
plt.subplot(2, 2, 1)
img=plt.imshow(I,cmap = 'gray')
plt.title('Original Image', fontsize = 16)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(magnitude,cmap = 'gray')
plt.title('Magnitude Response of Original Image', fontsize = 16)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(H,cmap = 'gray')
plt.title('Notch Reject Filter', fontsize = 16)
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(FI,cmap = 'gray')
plt.title('Filtered Image with Notch Reject Filter', fontsize = 16)
plt.axis('off')
fig.suptitle('Effect of Notch Reject Filter on Image', fontsize = 20)
plt.show()
```
