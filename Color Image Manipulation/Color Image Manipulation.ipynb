# <span style='background:black'><span style='color:Cyan'> <span style='font-family:"Times New Roman"'> Objective: 
### <span style='font-family:"Times New Roman"'>In this code, by modifying RGB channel of an image, we are trying to change the representation of targeted area in a color image. 

# <span style='font-family:"Times New Roman"'>Loading Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# <span style='font-family:"Times New Roman"'>Reading Files

```python
red = mpimg.imread('WashingtonDC-Band3-Red-564.tif')
green = mpimg.imread('WashingtonDC-Band2-Green-564.tif')
blue = mpimg.imread('WashingtonDC-Band1-Blue-564.tif')
nir = mpimg.imread('WashingtonDC-Band4-NearInfrared-564.tif')
```

# <span style='font-family:"Times New Roman"'>Creating RGB and NGB Images

```python
I_rgb = np.zeros((564,564,3))
I_ngb = np.zeros((564,564,3))
I_718 = np.zeros((564,564,3))
I_rgb[:,:,0] = red.astype(np.uint8)
I_rgb[:,:,1] = green.astype(np.uint8)
I_rgb[:,:,2] = blue.astype(np.uint8)
I_ngb[:,:,0] = nir.astype(np.uint8)
I_ngb[:,:,1] = green.astype(np.uint8)
I_ngb[:,:,2] = blue.astype(np.uint8)
```

# <span style='font-family:"Times New Roman"'>Displaying Images:

```python
fig = plt.figure(figsize=(14, 10))
plt.subplot(2, 3, 1)
img=plt.imshow(red,cmap = 'gray')
plt.title('Red Channel', fontsize = 16)
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(green,cmap = 'gray')
plt.title('Green Channel', fontsize = 16)
plt.axis('off')
plt.subplot(2, 3, 3)
plt.imshow(blue,cmap = 'gray')
plt.title('Blue Channel', fontsize = 16)
plt.axis('off')
plt.subplot(2, 3, 4)
plt.imshow(nir,cmap = 'gray')
plt.title('NearInfrared Image', fontsize = 16)
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow(I_rgb/255)
plt.title('RGB Image', fontsize = 16)
plt.axis('off')
plt.subplot(2, 3, 6)
plt.imshow(I_ngb/255)
plt.title('NI,GB Image', fontsize = 16)
plt.axis('off')
fig.suptitle('Multi Spectral Image of Washington D.C.', fontsize = 20)
plt.show()
```

```python
I_718 = I_ngb
I_718[(I_718[:,:,1]>120) & (I_718[:,:,2]>120)] = [255,255,0]
I_718[I_718[:,:,0]<20] = [0,210,255]
fig = plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.imshow(I_718/255)
plt.title('I_718 Image', fontsize = 16)
plt.axis('off')
I_ngb[:,:,0] = nir.astype(np.uint8)
I_ngb[:,:,1] = green.astype(np.uint8)
I_ngb[:,:,2] = blue.astype(np.uint8)
plt.subplot(1, 2, 2)
plt.imshow(I_ngb/255)
plt.title('NI,GB Image', fontsize = 16)
plt.axis('off')
fig.suptitle('Multi Spectral Image of Washington D.C.', fontsize = 20)
plt.show()
```
